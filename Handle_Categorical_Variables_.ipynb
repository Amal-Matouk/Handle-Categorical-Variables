```python
import pandas as pd

# initialize data
data = {'Subject': ['Math', 'Science', 'Math', 'Chemistry','Chemistry', 'Biology', 'Science', 'Math'],
        'Mark': [90, 60, 78, 55,34, 45, 67, 88]}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
df
```

```python
######## Using sklearn ########

#copy dataframe to new df to apply encoding 
df1=df.copy()
df1

from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
##encode
df1['Subject']=label_encoder.fit_transform(df1['Subject'])
df1
```

```python
#to reverse values to the original ones 
df1['Subject']=label_encoder.inverse_transform(df1['Subject'].ravel())
df1
```

```python
df2=df.copy()
df2

# Encoding Categorical Data using OneHotEncoder

from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
res= onehotencoder.fit_transform(df2[["Subject"]]).toarray()

dfn = df2.drop(columns='Subject').join(pd.DataFrame(res))
dfn
```

```python
#convert dataframe to array
x=df.to_numpy()

# Encoding Categorical Data using OneHotEncoder and ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
# creating one hot encoder object with categorical feature 0
# indicating the first column
columnTransformer = ColumnTransformer([('encoder',
                                        OneHotEncoder(),
                                        [0])],
                                      remainder='passthrough')
res = columnTransformer.fit_transform(x)


dfnew = pd.DataFrame(res)
dfnew

```
